```python
# making a connection to mongo db
import pymongo
import re
client= pymongo.MongoClient("mongodb+srv://krithi:krithi@cluster0.5r88k.mongodb.net/myFirstDatabase?retryWrites=true&w=majority")
db=client.user
records=db.books
```

```python
# validate phone book user details-name,phone number length is equal to 10
#email is validated with special character @ and .com in between
#through TRY and Exception

regex="^[a-zA-Z0-9-_]+@[a-zA-Z0-9]+.[a-z]{1,3}$"
a=int(input())
for i in range(a):
  name=list(input("enter name:").split())
  l=[]
  for x in name:
      l.append(x.capitalize())
  name=" ".join(l)
  phone=input("enter your number:")
  while len(phone)!=10 :
    phone=input("enter valid no")
  email=input("enter email")
  while not re.fullmatch(regex,email):
    email=input("enter valid")
  d = {
     "name1":name,
     "ph":phone,
     "mail":email
  }
  try:
    records.insert_one(d)  
    print("data inserted")
  except:
    print("failed")



```
